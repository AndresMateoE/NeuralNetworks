```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Test_conv_64 = np.loadtxt('test_loss_conv_64.csv')
Train_conv_64 = np.loadtxt('train_loss_conv_64.csv')

Test_lin_512 = np.loadtxt('test_loss_lineal_512.csv')
Train_lin_512 = np.loadtxt('train_loss_lineal_512.csv')

Test_conv_2 = np.loadtxt('test_loss_conv_2.csv')
Train_conv_2 = np.loadtxt('train_loss_conv_2.csv')

Test_lin_2 = np.loadtxt('test_loss_lin_2.csv')
Train_lin_2 = np.loadtxt('train_loss_lin_2.csv')
```

```python
x = np.arange(1,51)

plt.plot(x,Test_conv_64, label='Test Conv64',linestyle='-', c='red')
plt.plot(x,Train_conv_64, label='Train Conv64', linestyle='--', c='red')
plt.plot(x,Test_lin_512, label='Test Lin', linestyle='-', c='blue')
plt.plot(x,Train_lin_512, label='Train Lin', linestyle='--', c='blue')
plt.plot(x,Test_conv_2, label='Test Conv2', linestyle='-', c='green')
plt.plot(x,Train_conv_2, label='Train Conv2', linestyle='--', c='green')
#plt.plot(x,Test_lin_2, label='Test Lin2', linestyle='-', c='brown')
#plt.plot(x,Train_lin_2, label='Train Lin2', linestyle='--', c='brown')

plt.plot(plotsize=(12, 6))

plt.title("Comparaci√≥n Error Distintas Redes")
plt.ylabel("Loss")
plt.xlabel("Epoca")
plt.legend()
plt.savefig("Grafico_todos_Errores.png", dpi =300)
plt.show()
```
